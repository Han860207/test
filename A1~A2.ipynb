```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import os
from webdriver_manager.chrome import ChromeDriverManager
import re
import sqlite3
```

```python
chrome = webdriver.Chrome(ChromeDriverManager().install())
chrome.get('https://www.cwb.gov.tw/V8/C/C/Statistics/monthlymean.html')
soup = BeautifulSoup(chrome.page_source,"html.parser")
table =soup.find(class_='flex_table mobile_table-month')
row_and_col = table.find_all('th')
data = table.find_all('td')
chrome.close()
```

```python
def f(target) :
    columns = []
    for i in row_and_col :
        tmp = str(i)
        if target in tmp :
            m= re.findall(r'\>(\w+)\<',tmp)
            if m != [] :
                columns.append(m[0])
    return columns
columns = f('col')
```

```python
row= f('row')
memo=[]
df = pd.DataFrame(columns = columns)
for i in range(0,len(data)) :
    memo.append(data[i].text)
    if len(memo)% 14== 0 :
        sdf = pd.DataFrame([tuple(memo)],columns=columns)
        df = df.append(sdf,ignore_index=True)
        memo =[]
df.index = row
df
```

```python
new_df = pd.DataFrame(columns= ['month','area','temp'])
for i in range(len(df.columns)):
    for j in range(len(df.index)):
        new_df=new_df.append(pd.DataFrame([tuple([df.columns[i],df.index[j],df.iloc[j][i]])],columns= ['month','area','temp']))
new_df
```

```python
conn = sqlite3.connect('avg_temp.db')
cursor = conn.cursor()
new_df.to_sql('temp',conn,if_exists='append',index = True)
conn.commit()
conn.close()
```
